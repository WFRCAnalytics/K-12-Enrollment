```python
import pandas as pd
import matplotlib.ticker as mtick
import numpy as np
from arcgis.gis import *
gis = GIS()
```

```python
#Layer name for K-12 enrollment data
lyrK12 = "Utah_Schools_PreK_to_12"

sdfK12 = pd.DataFrame.spatial.from_featureclass(lyrK12)
sdfK12.columns
```

```python
#initial look at data
sdfK12
```

```python
sdfK12.groupby(['SchoolLeve','PrivateSch','CharterSch','OnlineScho'], as_index=False).agg(
                                                                                          K  =('K'       ,'sum'),
                                                                                          G01=('Grade1'  ,'sum'),
                                                                                          G02=('Grade2'  ,'sum'),
                                                                                          G03=('Grade3'  ,'sum'),
                                                                                          G04=('Grade4'  ,'sum'),
                                                                                          G05=('Grade5'  ,'sum'),
                                                                                          G06=('Grade6'  ,'sum'),
                                                                                          G07=('Grade7'  ,'sum'),
                                                                                          G08=('Grade8'  ,'sum'),
                                                                                          G09=('Grade9'  ,'sum'),
                                                                                          G10=('Grade10' ,'sum'),
                                                                                          G11=('Grade11' ,'sum'),
                                                                                          G12=('Grade12' ,'sum'),
                                                                                          Tot=('TotalK12','sum'))
```

```python
#created filtered dataset that removes unnecessary rows

#display total number of schools in original datset
display ("Total Schools: " + str(sdfK12.shape[0]))

sdfK12_filtered = sdfK12.copy()

#remove NONE and PREK schools, since no data
sdfK12_filtered = sdfK12_filtered[(sdfK12_filtered['SchoolLeve'] != 'PREK') & (sdfK12_filtered['SchoolLeve'] != 'NONE')]
display ("Total Schools After NONE/PREK Removed: " + str(sdfK12_filtered.shape[0]))

#remove online schools
sdfK12_filtered = sdfK12_filtered[sdfK12_filtered['OnlineScho'] != 'Y']
display ("Total Schools After Online Removed: " + str(sdfK12_filtered.shape[0]))

#inlcude only desired columns
sdfK12_filtered = sdfK12_filtered[['FID','OBJECTID', 'LEAName', 'SchoolName',
                                   'CharterSch', 'PrivateSch', 'OnlineScho',
                                   'GradeLow', 'GradeHigh', 'SchoolLeve', 'SchoolType', 'TotalK12', 'K',
                                   'Grade1', 'Grade2', 'Grade3', 'Grade4', 'Grade5', 'Grade6', 'Grade7',
                                   'Grade8', 'Grade9', 'Grade10', 'Grade11', 'Grade12','SHAPE']]



display ("Total Filtered Enrollment: " + str(sdfK12_filtered['TotalK12'].sum()))

display(sdfK12_filtered)
```

```python
#Set private and charter schools to PriCha group
sdfK12_filtered.loc[((sdfK12_filtered['PrivateSch']=='Y')),'Group'] = 'PriCha'
sdfK12_filtered.loc[((sdfK12_filtered['CharterSch']=='Y')),'Group'] = 'PriCha'
sdfK12_filtered.loc[((sdfK12_filtered['PrivateSch']!='Y') & (sdfK12_filtered['CharterSch']!='Y')),'Group'] = 'Public'

#add new Level column
sdfK12_filtered['School_Level'] = sdfK12_filtered['SchoolLeve'] 

sdfK12_filtered.groupby(['Group'], as_index=False).agg(NumSchools=('TotalK12','size'))
```

```python
gradesums = sdfK12_filtered.groupby(['School_Level','Group'], as_index=False).agg(K  =('K'      ,'sum'),
                                                                       G01=('Grade1' ,'sum'),
                                                                       G02=('Grade2' ,'sum'),
                                                                       G03=('Grade3' ,'sum'),
                                                                       G04=('Grade4' ,'sum'),
                                                                       G05=('Grade5' ,'sum'),
                                                                       G06=('Grade6' ,'sum'),
                                                                       G07=('Grade7' ,'sum'),
                                                                       G08=('Grade8' ,'sum'),
                                                                       G09=('Grade9' ,'sum'),
                                                                       G10=('Grade10','sum'),
                                                                       G11=('Grade11','sum'),
                                                                       G12=('Grade12','sum'),
                                                                       Tot=('TotalK12','sum'))

display(gradesums)
```

Most ELEM schools with Grade 8 enrollment > 0, stop at Grade 8 and have very little 9-12 enrollment. So, will set category for all ELEM schools with Grade 8 enrollment > 0 as K8.

```python
#join to get enrollment levels
sdfK12_enrolllevels = pd.DataFrame.merge(sdfK12_melt,dfEnrollLevels,on=('School_Level','Grade'))
sdfK12_enrolllevels = sdfK12_enrolllevels.groupby(['FID','Group','Enrol_Levl'],as_index=False).agg(Enrol_Levl_Total=('Enrollment','sum'))
sdfK12_enrolllevels
```

```python
sdfK12_enrolllevels.dtypes
```

```python
sdfK12_enrol = sdfK12_enrolllevels.pivot(index=['FID'], columns=['Group','Enrol_Levl'],values='Enrol_Levl_Total')

#collapse multi-index
sdfK12_enrol.columns = ['_'.join(col).strip() for col in sdfK12_enrol.columns.values]

sdfK12_enrol = sdfK12_enrol.fillna(0)

sdfK12_enrol = sdfK12_enrol.astype(int)

sdfK12_enrol = sdfK12_enrol[['Public_Elem','Public_Midl','Public_High','PriCha_Elem','PriCha_Midl','PriCha_High']]
sdfK12_enrol
```
